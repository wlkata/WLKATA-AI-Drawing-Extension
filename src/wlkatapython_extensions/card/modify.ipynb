```python
from pathlib import Path
```

```python
label_root = Path("datasets/poker/labels")
```

```python
for subset in ["train", "test", "val"]:
    for label_file in label_root.joinpath(subset).iterdir():
        if label_file.suffix == ".txt":
            with open(label_file, "r") as f:
                label = f.read()
            print(label_file.name)
            splits = label.split(' ')
            print(splits[0])
            if int(splits[0]) == 52:
                continue
            new_id = index_mapping[int(splits[0])]
            print(f"{new_id} {' '.join(splits[1:])}")
            with open(label_file, "w") as f:
                f.write(f"{new_id} {' '.join(splits[1:])}")
```
